# Merge aircraft data into test data

```python
import pandas as pd
```

```python
# import both datasets as pandas DataFrame
ac = pd.read_csv('aircraft.csv', index_col=0)
train = pd.read_csv('train_flights_Jan_1w.csv')
```

```python
ac.head()
```

```python
train.head()
```

```python
# process merge
train = train.merge(ac, how='left', left_on='tail_num', right_on='registration')
```

```python
# drop redundant columns
train.drop(['registration'], axis=1, inplace=True)
```
