# Merge airpot data into test data

```python
import pandas as pd
```

```python
# import both datasets as pandas DataFrame
astrike = pd.read_csv('airports_strikes.csv', index_col=0)
train = pd.read_csv('train_flights_Jan_1w.csv')
```

```python
astrike.head()
```

```python
train.head()
```

```python
# process merge, drop duplicate column
train = train.merge(astrike, how='left', left_on='origin', right_on='IATA')
train.drop(['IATA'], axis=1, inplace=True)
```
