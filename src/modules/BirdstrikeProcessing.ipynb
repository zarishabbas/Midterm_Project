### Count of incidents by airport_id, date

```python
byday = df[['INDX_NR', 'INCIDENT_DATE ', 'AIRPORT_ID']]
```

```python
byday = byday.groupby(['AIRPORT_ID', 'INCIDENT_DATE ']).count().reset_index()
```

```python
byday.rename(columns={'INDX_NR': 'COUNT'}, inplace=True)
```

```python
byday.head()
```

### Incidents by flight no.

```python
byflight = df[['FLT', 'OPID', 'INCIDENT_DATE ', 'TIME', 'TIME_OF_DAY', 'AIRPORT_ID', 'LATITUDE', 'LONGITUDE', 'FAAREGION']]
```

```python
byflight
```

### Merge data

```python
data = pd.merge(byflight, byday,  how='left', left_on=['AIRPORT_ID','INCIDENT_DATE '], right_on = ['AIRPORT_ID','INCIDENT_DATE '])
```

```python
data
```

```python
data.to_csv('birdstrike.csv')
```
